```python
df['Churn'].value_counts()
```

```python
df['Churn'].value_counts(normalize=True)
```

*Approximately 26.53% of customers have churned, while 73.46% have retained the service. This indicates a moderate churn rate, highlighting the importance of identifying factors contributing to customer attrition.*

🔹**Gender vs Churn**

❓ **Question**

Does gender influence customer churn?

```python
df.groupby('gender')['Churn'].value_counts(normalize=True).unstack()
```

📊 **Insight**

The churn rate across genders shows minimal variation.

🧠 **Interpretation**

This suggests that gender does not significantly impact customer retention behavior.

✅ **Conclusion**

Gender is a weak predictor of churn and does not play a major role in customer attrition.

🔹**Contract vs Churn**

❓ **Question**

Does contract type influence churn?

```python
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
```

📊 **Insight**

Customers with month-to-month contracts have significantly higher churn rates compared to those with long-term contracts.

🧠 **Interpretation**

This indicates that customers with short-term commitments are more likely to discontinue the service.

✅ **Conclusion**

Contract type is a strong factor influencing churn, with long-term contracts improving customer retention.

🔹**Tenure vs Churn**

❓ **Question**

Does tenure influence churn?


```python
df.groupby('Churn')['tenure'].mean()
```

📊 **Insight**

Customers who churn have lower average tenure compared to those who remain.

🧠 **Interpretation**

This suggests that newer customers are more likely to discontinue the service.

✅ **Conclusion**

Tenure plays a significant role in churn, with shorter tenure associated with higher churn rates.

🔹**Monthly Charges vs Churn**

❓ **Question**

Does Monthly Charges influence churn?

```python
df.groupby('Churn')['MonthlyCharges'].mean()
```

📊 **Insight**

Customers who churn have higher average monthly charges compared to those who remain.

🧠 **Interpretation**

This suggests that higher pricing may contribute to customer dissatisfaction, increasing the likelihood of churn.

✅ **Conclusion**

Monthly charges appear to influence churn behavior, indicating that pricing strategy may play a role in customer retention.

🔹 **Final Observations**

- Gender shows negligible difference in churn, indicating it is not a significant factor.
- Contract analysis indicates that customers with long-term contracts have lower churn compared to month-to-month customers, making it a key factor.
- Tenure analysis shows that customers with shorter tenure are more likely to churn, while long-term customers tend to stay, making tenure an important factor.
- Monthly charges analysis indicates that higher charges are associated with increased churn, making it a moderate influencing factor.


Overall, churn is primarily driven by contract type, tenure, and pricing factors, while demographic attributes such as gender have minimal impact.
